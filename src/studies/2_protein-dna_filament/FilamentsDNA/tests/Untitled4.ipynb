```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_csv(r'/Users/thor/Downloads/Balkon_renovatie_vve_vrolike_noot_22.csv')
df = df_raw[:27]
notes = df_raw[30:]
df = df.apply(lambda x: x.str.replace(',','.'))
df = df.sort_values('datum')
# convert "Fee" from string to float
df['kosten'] = df['kosten'].astype(float)
df['datum'] = df['datum'].astype(int)
```

```python
fig,ax=plt.subplots(figsize=[6,3])

df_balkons = df.loc[df['werk'] == 'balkon']
df_balkons['cumsum'] = list(np.cumsum(list(df_balkons['kosten'])))

x = list(df_balkons['datum'])
y = list(df_balkons['cumsum'])

ax.plot(x,y)
ax.scatter(x,y)

df_meerwerk = df.loc[df['werk'] == 'meerwerk']
df_meerwerk['cumsum'] = list(np.cumsum(list(df_meerwerk['kosten'])))

xx = list(df_meerwerk['datum'])
yy = list(df_meerwerk['cumsum'])

from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='tab:blue', lw=4),
                Line2D([0], [0], color='tab:orange', lw=4)]



ax.legend(custom_lines, ['Begroot', 'Meerwerk'])

ax.plot(xx,yy)
ax.scatter(xx,yy)

ax.set_title(f'Begroot {round(y[-1])}, Meerwerk {round(yy[-1])}, Totaal {round(y[-1]+yy[-1])}')
ax.set_ylabel('Cumulatieve kosten [euro]')
ax.set_xlabel('Datum')

fig.savefig(f'begroot_v_meerwerk_bakons_22.png',dpi=200, bbox_inches='tight')
```
