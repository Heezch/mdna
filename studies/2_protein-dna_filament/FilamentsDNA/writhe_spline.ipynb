The writhe (Wr) of a curve, in the context of DNA or any filament, can be thought of as a measure of the spatial configuration of the curve in three dimensions. It quantifies how much the curve deviates from a planar circle or a straight line. For a given integer value of writhe, we want to construct a curve that loops around itself that number of times.

To construct such a curve, we can use a parametric representation of a three-dimensional curve that has a given number of crossings or self-intersections when projected onto a plane. Here's how we can approach this:

- Base Curve: Start with a simple parametric curve, like a circle or helix, defined in 3D space.

- Adding Loops: Introduce a modulation to the curve that adds loops. We can use trigonometric functions to create these loops. The number of peaks in the trigonometric function can correspond to the number of loops.

- Avoiding Self-Intersection: Ensure that the loops do not self-intersect. This might require adjusting the phase or amplitude of the modulation.

- Controlling Writhe: The number of loops added will be related to the writhe. Since we're not considering the sign of the writhe, we'll focus on the absolute value, which corresponds to the number of loops.

- Parametric Equations: Use a combination of sine and cosine functions for the x, y, and z coordinates, with parameters adjusted to control the number of loops.

Let's define a Python function to generate such a curve. We'll use a helical base with additional sinusoidal modulations to introduce the loops. The parameter n will control the number of loops, which corresponds to the absolute value of the writhe we want to model.

```python
# Set the parameter t which controls the density of the plot
t = np.linspace(-np.pi, np.pi, 1000)

# Define the parametric functions for each curve
def even_loop(t,n):
    return np.sin(t), np.sin(n*t) / n
def uneven_loop(t,n):
    return np.sin(t), np.sin(n*t) / n
n_loops = 8
# Create the plots for each loop function
fig, ax = plt.subplots(1, n_loops, figsize=(n_loops*4, 4))

for _,idx in enumerate(range(1,n_loops+1)):
    if idx % 2 == 0:
        loop = even_loop
    else:
        loop = uneven_loop
        
    x, y = loop(t, idx)
    ax[_].plot(x, y)
    ax[_].axis('equal')
```

```python
%matplotlib ipympl 
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

def plot_torus_knot(p, q, num_points=1000):
    """
    Plot a torus knot in 3D.
    
    :param p: integer, number of times that the knot will wind around the axis of symmetry
    :param q: integer, number of times that the knot will wind around a circle in the interior of the torus
    :param num_points: number of points to plot for the knot
    """
    theta = np.linspace(0, 2 * np.pi, num_points)
    
    # Parametric equations for a torus knot
    R = 2
    r = 1
    x = (R + r * np.cos(q * theta)) * np.cos(p * theta)
    y = (R + r * np.cos(q * theta)) * np.sin(p * theta)
    z = r * np.sin(q * theta)

    # Plotting
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z)
    ax.set_title(f'Torus Knot with p={p} and q={q}')
    plt.show()

# Example torus knot
plot_torus_knot(p=6, q=1)
```

```python
def plot_supercoiled_structure(num_turns=1, coils_per_turn=4, supercoil_density=5, num_points=1000):
    """
    Plot a supercoiled structure similar to an elastic rubber band.

    :param num_turns: integer, number of turns in the helix
    :param coils_per_turn: integer, number of coils in each turn
    :param supercoil_density: integer, density of supercoiling
    :param num_points: number of points to plot for the structure
    """
    t = np.linspace(0, num_turns * 2 * np.pi, num_points)

    # Helix parameters
    a = 10  # radius of the helix
    b = 0.1  # radius of the supercoil

    # Parametric equations for the supercoiled structure
    x = (a + b * np.cos(supercoil_density * t)) * np.cos(coils_per_turn * t)
    y = (a + b * np.cos(supercoil_density * t)) * np.sin(coils_per_turn * t)
    z = b * np.sin(supercoil_density * t)

    # Plotting
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x, y, z)
    ax.set_title('Supercoiled Structure')
    plt.show()

# Example of a supercoiled structure
plot_supercoiled_structure()
```
