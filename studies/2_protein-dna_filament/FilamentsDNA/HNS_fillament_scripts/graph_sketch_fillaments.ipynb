```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def add_DNA(G,length,circular=False):
    sequence = np.arange(0,length)
    connections = [[i,i+1] for i in sequence][:sequence[-1]]
    if circular:
        connections.append([0,sequence[-1]])
    G.add_nodes_from(sequence) 
    G.add_edges_from(connections)

def add_HNS(G,link,idx):
    G.add_node(f'{idx}_s1')
    G.add_node(f'{idx}_s2')
    G.add_node(f'{idx}_DBD')
    
    G.add_edge(f'{idx}_s1',f'{idx}_s2')
    G.add_edge(f'{idx}_s2',f'{idx}_DBD')
```

```python
G = nx.Graph()


add_DNA(G,20,circular=False)
add_HNS(G,None,0)
nx.draw(G)
```

```python
add_DNA(G,200,circular=True)

for i in range(0,100):
    add_HNS(G,None,i)
    
nx.draw(G,node_size=2)
```
